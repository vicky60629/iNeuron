### We have the min and max temperatures in a city In India for each months of the year.    <br>                                                   We would like to find a function to describe this and show it graphically, the dataset      <br>                                             given below.              <br>                                                                                                                                                            Task:                        <br>                                                                                                                                          1.fitting it to the periodic function        <br>                                                                                                                                                   2.plot the fit  <br>                                                                                                                                                                          Data  <br>                                                                                                                                                                             Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25  <br>                                                                                                             Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
temp_max=np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
temp_min=np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
month=np.arange(1,13)
plt.plot(month,temp_min,color='b',linestyle='-',marker='o')
plt.plot(month,temp_max,color='r',linestyle='-',marker='o')
plt.show()
```

```python
titanic= pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv')
titanic
```

```python
label = ['male','female']
count= titanic.sex.value_counts()
fig,a= plt.subplots()
a.pie(count, labels=label, autopct='%1.1f%%', colors = ['Blue','Red'])
plt.show()
```

```python
category= titanic[titanic.sex=='male'].plot.scatter('age', 'fare', color='blue',label='male')
titanic[titanic.sex=='female'].plot.scatter('age', 'fare',color='red',label='female',ax=category)
plt.show()
```
